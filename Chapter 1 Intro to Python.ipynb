## Write a python program for Armstrong number

```python
num = 121
a = num
sum = 0
while a > 0:
    digit = a % 10
    sum = sum + digit**3
    a //= 10
if sum == num:
    print("It is an Armstrong number")
else:
    print("It is not an Armstrong number")
```

```python
num = int(input("Enter a number: "))
n = num
power = len(str(num))
total = 0

while n > 0:
    digit = n % 10
    total += digit ** power
    n //= 10

if total == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```
