#### 1) Stable Top-K Selection
#### The function should return the top k elements based on score.

#### Input
#### items: List of tuples (name, score)
#### k: Integer


```python
def top_k(items, k):
    
    items.sort(key=lambda x: x[1], reverse=True)
    return items[:k]
```

#### 2) Category Word Counter

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        if cat not in d:
            d[cat] = {}
        if word not in d[cat]:
            d[cat][word] = 0
        d[cat][word] =+ 1
    return d
```

#### 3) Missing Number (0 to n)

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) / 2
    return total - sum(nums)

```

#### 4) Default Argument Bug


```python
def add_unique(x, seen = set()):
    current = set()
    current.clear()
    if x not in seen:
        seen.add(x)
        current.add(x)
    return list(current)
print(add_unique(1))
print(add_unique(2))
```

#### 5) Moving Average of Last w Elements


```python
def moving_average(stream, w):
    average = []
    sum = 0

    for i in range(len(stream)):
        sum += stream[i]

        if i >= w:
            sum -= stream[i -w]
        average.append(sum / min(i + i, w))
    return average
```

#### 7) Longest Unique Substring

```python
def longest_unique_substring(s):
    seen=[]
    start = 0
    max_len = 0

    for i, ch in enumerate(s):
        if ch in seen and seen[ch] >= start:
            start = seen[ch] + 1
        seen[ch] = i
        max_len = max(max_len, i - start + 1)
    return max_len
```
